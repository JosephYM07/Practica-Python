# Manejo de archivos

```python
def crear_txt():
    archivo=open('datos.txt','w')
    archivo.close()
def grabar_txt():
    archivo=open('datos.txt','a')
    archivo.write('Linea 1\n')
    archivo.write('Linea 2\n')
    archivo.write('Linea 3\n')
crear_txt()
grabar_txt()
```

```python
def leer_txt():
    archivo=open('datos.txt','r',encoding='utf-8')
    linea=archivo.readline()
    while linea!="":
        print(linea)
        linea=archivo.readline()
    archivo.close()
        
def leer_txt_en_lista():
    archivo=open('datos.txt','r',encoding='utf-8')
    lineas=archivo.readlines()
    print(lineas)
    archivo.close()
```

```python
leer_txt()
```
