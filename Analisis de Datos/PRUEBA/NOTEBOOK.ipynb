```python
archivo = open('17_LEYES_DEL_TRABAJO_EN_EQUIPO.txt', 'r', encoding='ANSI') # Abrir el archivo en modo lectura y codificación utf-8
mensaje = archivo.read().lower()  # Convertir a minúsculas
archivo.close() # Cerrar el archivo
```

```python
palabras = mensaje.split()  # Separar el texto en palabras
```

```python
frecuencia_palabras = {} # Crear un diccionario vacío
```

```python
for palabra in palabras:
    if palabra in frecuencia_palabras:
        frecuencia_palabras[palabra] += 1 # Si la palabra ya existe, aumentar el contador 
    else:
        frecuencia_palabras[palabra] = 1 # Si la palabra no existe, inicializar su contador en 1 y se agrega al diccionario
```

```python
palabras_ordenadas = sorted(frecuencia_palabras.items()) # El método items() devuelve una lista de tuplas y sorted ordena la lista alfabéticamente por la primera posición de la tupla
```

```python
# Imprimir los resultados
print("LISTADO DE PALABRAS")
listado = "" # Crear una cadena vacía
for palabra, frecuencia in palabras_ordenadas: # Recorrer la lista de tuplas
    listado += f"[{palabra} - {frecuencia}]\n " # Concatenar la cadena

print(listado) # Imprimir la cadena
Sumatoria=sum(frecuencia_palabras.values())
print("Sumatoria",Sumatoria)
```
